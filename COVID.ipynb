# COVID Data Visualization

## Importing Pandas and Matplotlib

```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.pyplot import subplots, show
```

## Reading CSV

```python
df = pd.read_csv('Output_vaccination.csv')
```

## Visualizing Vaccinations vs New Cases in India

```python
df1 = df[df['location'] == 'India']
```

```python
df1['total_vaccinations'] = df1['total_vaccinations'].fillna(0)
```

```python
fig = plt.figure(figsize=(20,5))
ax1 = fig.add_subplot(121)
ax1.xaxis.set_visible(False)
ax2 = fig.add_subplot(122)
ax2.xaxis.set_visible(False)
ax1.plot(df1['date'],df1['total_vaccinations'])
ax2.plot(df1['date'],df1['new_cases'],color = 'red')
ax1.set_title('Vaccination over time')
ax2.set_title('New Cases over time')
fig.text(0.5, 0.04, 'Vaccinations vs New Cases of COVID Trend in India', ha='center', va='center',fontsize=15)
show()
```
